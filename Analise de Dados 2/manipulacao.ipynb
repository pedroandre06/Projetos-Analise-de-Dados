Importação Das Bibliotecas

```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt




```

Puxando e lendo a base de Dados

```python
df = pd.read_csv("vendas.csv")

display(df)
```

Descrição da tabela

```python
describe = df.describe().round(2)
display(describe) 
```

Verificando Valores Vazios 


```python
vazio = df.isnull().sum()
display(vazio)
```

Total de Preço Unitário

```python
total = df["preco_unitario"].sum()
print (f"R${total:,.2f}")
```

```python

produtos_mais_vendidos = df.groupby("produto")["quantidade"].sum().sort_values(ascending=False)

# Gráfico de barras
produtos_mais_vendidos.plot(kind="bar", figsize=(10,5))
plt.title("Produtos mais vendidos (por quantidade)")
plt.xlabel("Produto")
plt.ylabel("Quantidade Vendida")
plt.xticks(rotation=80)
plt.show()
```
